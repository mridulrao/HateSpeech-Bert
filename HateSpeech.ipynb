```python
import datasets 
dataset = datasets.load_dataset('ucberkeley-dlab/measuring-hate-speech', 'binary')   
df = dataset['train'].to_pandas()
df.describe()
```

```python
df['text']
```

```python
import matplotlib.pyplot as plt

plt.bar(df['hate_speech_score'])
```

```python
df['hate_speech_score'].min()
```

```python
intervals = {}
interval_width = 2.0

for value in df['hate_speech_score']:
    interval = int(value / interval_width) * interval_width
    if interval in intervals:
        intervals[interval] += 1
    else:
        intervals[interval] = 1

# Print the results
for interval, count in intervals.items():
    print(f"Interval {interval} to {interval + interval_width}: {count} values")
```

```python
'''
hate_speech = []
for i in range(len(df['text'])):
    if df['hate_speech_score'].iloc[i] > 2:
        hate_speech.append(df['text'].iloc[i])
'''

hate_speech = df[df['hate_speech_score'] > 2]
df_hate = hate_speech[['hate_speech_score', 'text']]
```

```python
df_hate
```

```python
not_hate_speech = df[df['hate_speech_score'] < -3]
df_not_hate = not_hate_speech[['hate_speech_score', 'text']]
```

```python
df_not_hate
```

```python
df_hate['hate'] = 1
df_not_hate['hate'] = 0
```

```python
df_not_hate
```

```python
import pandas as pd

df = pd.concat([df_hate, df_not_hate], axis=0, ignore_index=True)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2)
```

```python
df_bert = pd.DataFrame({
    'id':range(len(df_train)),
    'label':df_train['hate'],
    'extra_col':['a']*df_train.shape[0],
    'text': df_train['text']
})


df_bert
```

```python
df_bert_train, df_bert_dev = train_test_split(df_bert, test_size=0.01)
```

```python
df_bert_test = pd.DataFrame({
    'id':range(len(df_test)),
    'text': df_test['text']
})

df_bert_test_labels = pd.DataFrame({
    'id':range(len(df_test)),
    'text': df_test['text'],
    'label' : df_test['hate']
})

df_bert_test.head()


df_bert_test
```

```python
# Saving dataframes to .tsv format as required by BERT
df_bert_train.to_csv('data/train.tsv', sep='\t', index=False, header=False)
df_bert_dev.to_csv('data/dev.tsv', sep='\t', index=False, header=False)
df_bert_test.to_csv('data/test.tsv', sep='\t', index=False, header=False)
df_bert_test_labels.to_csv('data/test_labels.tsv', sep='\t', index = False, header = False)
```
